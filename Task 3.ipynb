```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
movies.head()
```

```python
movies.shape
movies.info()
```

```python
import matplotlib.pyplot as plt

# 1. Budget vs Revenue
plt.figure(figsize=(8, 5))
plt.scatter(movies['budget'], movies['revenue'], alpha=0.5)
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
#Top 10 Most Popular Movies
top_10_popular_movies = movies.sort_values(by='popularity', ascending=False).head(10)
plt.figure(figsize=(8, 5))
plt.barh(top_10_popular_movies['title'], top_10_popular_movies['popularity'], color='skyblue')
plt.title('Top 10 Most Popular Movies')
plt.xlabel('Popularity')
plt.ylabel('Movie Title')
plt.gca().invert_yaxis()
plt.show()
```

```python
#Top 10 Movies by Revenue
top_10_revenue_movies = movies.sort_values(by='revenue', ascending=False).head(10)
plt.figure(figsize=(8, 5))
plt.barh(top_10_revenue_movies['title'], top_10_revenue_movies['revenue'], color='skyblue')
plt.title('Top 10 Movies by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Movie Title')
plt.gca().invert_yaxis()
plt.show()
```

```python
#  Top 10 Movies by Budget
top_10_budget_movies = movies.sort_values(by='budget', ascending=False).head(10)
plt.figure(figsize=(8, 5))
plt.barh(top_10_budget_movies['title'], top_10_budget_movies['budget'], color='skyblue')
plt.title('Top 10 Movies by Budget')
plt.xlabel('Budget')
plt.ylabel('Movie Title')
plt.gca().invert_yaxis()
plt.show()
```

```python
#  Top 10 Movies by Vote Count
top_10_vote_count_movies = movies.sort_values(by='vote_count', ascending=False).head(10)
plt.figure(figsize=(8, 5))
plt.barh(top_10_vote_count_movies['title'], top_10_vote_count_movies['vote_count'], color='skyblue')
plt.title('Top 10 Movies by Vote Count')
plt.xlabel('Vote Count')
plt.ylabel('Movie Title')
plt.gca().invert_yaxis()
plt.show()
```
